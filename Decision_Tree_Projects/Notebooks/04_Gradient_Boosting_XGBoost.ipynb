## Gradient Boosting - XGBoost

### Importing the Libraries and Dataset

XGBoost is used when: ✔ Tabular data ✔ Non-linear relationships ✔ Need top performance ✔ Interpretability via SHAP

Avoid it when: ❌ Dataset is tiny ❌ Model transparency must be absolute ❌ Training time is extremely constrained


```python
import pandas as pd

df_encoded = pd.read_pickle("churn_data_cleaned.pkl")
df_encoded.head(10)
```

### Splitting the data into Training and Testing sets.

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop(columns=['Churn'])
y = df_encoded['Churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

### Creating the Baseline Gradient Boost.

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report

gb_baseline = GradientBoostingClassifier(random_state=42)
gb_baseline.fit(X_train, y_train)

print(classification_report(
    y_test,
    gb_baseline.predict(X_test)
))
```

### Gradient Boost Tuning

```python
gb_tuned = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    random_state=42
)

gb_tuned.fit(X_train, y_train)

print(classification_report(
    y_test,
    gb_tuned.predict(X_test)
))
```

### Introducing XGBoost.

```python
import xgboost
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss',
    random_state=42
)

xgb.fit(X_train, y_train)

print(classification_report(
    y_test,
    xgb.predict(X_test)
))
```
